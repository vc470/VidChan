### Cleaning

```python
wb_clean = wb_data_raw.copy()
wb_clean.info()
```

```python
# Creating dataframe of Missing Years Counts for each country 

cols = list(wb_clean.columns[3:])

myc = wb_clean[wb_clean['gdp_per_cap'].isna()].groupby('country').count()
myc.reset_index(inplace=True)
myc = myc.loc[:, ['country','year']]
myc.rename({'year':'gdp_per_cap'}, axis=1, inplace=True)

for col in cols:
    myc_col = wb_clean[wb_clean[col].isna()].groupby('country').count()
    myc_col.reset_index(inplace=True)
    myc_col = myc_col.loc[:, ['country','year']]
    myc_col.rename({'year':col}, axis=1, inplace=True)
    myc = pd.merge(left=myc, right=myc_col, how='outer',
                  left_on='country', right_on='country')

myc.head()
# Example: Afghanistan is missing 12 years for the gdp_per_cap column
# but isn't missing any years for the mobilesub_per100peeps column
```

```python
# count is number of countries missing at least 1 year of that data
myc.describe()

# I'm going to use a high count and a high mean/median to determine
# if some columns should be totally removed.

# For instance, agg.empl.agri.perce has a high count, but a low mean/median.
# That means a lot of countries are missing that data, but only a few years
# on average. That column should remain and just have some data imputed.

# On the other hand, Manufacturing... has a high count and a high mean/median.
# That means many countries are missing a significant number of years.
# I'm going to remove that column.

# intl_tourist_arrival is debatable. I'm going to leave it for now,
# but it could be removed if we find it isn't useful anyway.
```

```python
# Removing Manufacturing... from both myc and wb_clean
myc.drop('Manufacturing_added_annual%growth', axis=1, inplace=True)
wb_clean.drop('Manufacturing_added_annual%growth', axis=1, inplace=True)
myc.head()
```

```python
# 11 columns, 27 years each = 297 possible missing years for a country

# I'm going to drop countries that are missing more than 60 total years
# (about 20% of 297)

# Makes a column for the total missing years for each country
cols_to_sum = list(myc.columns[1:])
myc['total_missing_years'] = myc[cols_to_sum].sum(axis=1)

myc[['country', 'total_missing_years']].head()
```

```python
# Removing countries missing more than 60 total years
drop_countries = myc.loc[myc['total_missing_years'] > 60, 'country'].tolist()

for country in drop_countries:
    indexNames = wb_clean[wb_clean['country'] == country].index
    wb_clean.drop(indexNames, inplace=True)

wb_clean.reset_index(drop=True, inplace=True)
wb_clean.head()
```

```python
# Going to remake the myc
cols = list(wb_clean.columns[3:])


myc = wb_clean[wb_clean['gdp_per_cap'].isna()].groupby('country').count()
myc.reset_index(inplace=True)
myc = myc.loc[:, ['country','year']]
myc.rename({'year':'gdp_per_cap'}, axis=1, inplace=True)

for col in cols:
    myc_col = wb_clean[wb_clean[col].isna()].groupby('country').count()
    myc_col.reset_index(inplace=True)
    myc_col = myc_col.loc[:, ['country','year']]
    myc_col.rename({'year':col}, axis=1, inplace=True)
    myc = pd.merge(left=myc, right=myc_col, how='outer',
                  left_on='country', right_on='country')

myc.head()
```

### Imputing Data

```python
# Notice missing values in agriculture and tourism columns
# for Albania, which we can confirm are fixed below
wb_clean.head()
```

```python
# I'm going to fill in the missing data (grouped by country) by
# defining a function based on interpolate.

# This function takes one column of the dataframe and fills
# in missing values by country (unless the country is missing all of them).
def impute_func(df, col):
    # List of countries to loop through
    countries = sorted(list(set(df['country'])))
    
    # Initializing master df to add results to
    imputed_df = pd.DataFrame(columns=['country', 'year', col])

    for country in countries:
        # Set up dataframe for the one country
        country_df = df.loc[df['country']==country, ['country','year',col]].copy()
        
        # Convert the Year column to DateTime format and set it as index
        # (requirement for interpolate)
        country_df['year'] = pd.to_datetime(country_df['year'], format='%Y')
        country_df = country_df.set_index('year')
    
        # Use interpolate on the given column with the time method
        # to fill in missing values.
        country_df[col] = country_df[col].interpolate(method='time', limit_direction='both')
        
        # Change Year back to an int column
        country_df.reset_index(inplace=True)
        country_df['year'] = country_df['year'].dt.year
        
        # Add specific country df to the master df
        imputed_df = pd.concat([imputed_df, country_df], axis=0)
        
    return imputed_df
```

```python
# Now I need to use that function on all the columns of the WB df

# List of columns other than country and year
cols = list(wb_clean.columns[2:])

# Initializing master df to add the imputed columns to
imputed_wb = wb_clean[['country', 'year']].copy()

# For loop imputes each column and adds it to the master df
for col in cols:
    imputed_col = impute_func(wb_clean, col)
    imputed_wb = pd.merge(left=imputed_wb, right=imputed_col, how='right',
                  left_on=['country', 'year'], right_on=['country', 'year'])


imputed_wb.head()

# If missing values aren't inbetween two available values,
# they are just filled in with the next closest available value
# (that's why 36.41 is repeated for the first part of agri_perc_gdp)
```

```python
# There are still a few missing values due to some
# countries missing all of the years for a column
imputed_wb.info()
```

```python
# I'm going to drop those countries entirely
wb_final = imputed_wb.dropna()

# Looking at which countries were dropped (all pretty small)
imputed_countries = set(imputed_wb['country'])
final_countries = set(wb_final['country'])
imputed_countries - final_countries
```
